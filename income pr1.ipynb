```python
# actual data has ' ?' is null values

data = data.replace(' ?', np.nan)

round((data.isnull().sum()/data.shape[0])*100, 2).astype('str') + ' %'
```

```python
# 15 num columns had ? earlier which is replaced by NaN

data.iloc[14]
```

```python
data = data.replace(np.nan, '?')
# this code give description of % of null values in data

columns = data.columns
datapoints = data.shape[0]

for i in columns:
    num_null = data[i].isin(['?']).sum()
    if num_null > 0:
        print(i)
        print(num_null)
        print("{0:.2f}".format(float(num_null)/datapoints * 100) +' %' )
```

# processing categorical features

```python
# these categorical features should be transformed.

# label encoding

from sklearn.preprocessing import LabelEncoder


for col in data.columns:
    if data[col].dtype == 'object':
        encoder = LabelEncoder()
        data[col] = encoder.fit_transform(data[col])
        
data
```
